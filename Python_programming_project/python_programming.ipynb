```python
import sqlite3

#extract the list of dictionaries from your JSON data
json_data_list= data['objects']

#create an sqlite database names 'data.db'
conn = sqlite3.connect('data.db')
cursor= conn.cursor()

#create the 'employee' table with appropriate column names
create_table_sql="CREATE TABLE IF NOT EXISTS employee (ID TEXT, JobTitle TEXT, EmailAddress TEXT, FirstNameLastName TEXT, vaccinated TEXT)"
cursor.execute(create_table_sql)

#insert data from dictionaries in JSON into the 'employee' table
for item in json_data_list:
  values= (item.get('ID', ''), item.get('JobTitle', ''), item.get('EmailAddress', ''), item.get('FirstNameLastName', ''), item.get('vaccinated', ''))
  insert_sql= 'INSERT INTO employee VALUES (?, ?, ?, ?, ?);'
  cursor.execute(insert_sql, values)

conn.commit()
conn.close()
```

```python
import sqlite3

#read data from sqlite database into pandas dataframe
conn= sqlite3.connect('data.db')
query="SELECT * FROM employee"
df=pd.read_sql_query(query, conn)

#define a function for the vaccinated employee
def calculate_bonus(vaccinated):
  return 15 if vaccinated == 'True' else 5

# apply the function to create the "bonus_percent" column
df['bonus_percent']=df['vaccinated'].apply(calculate_bonus)

#export the dataframe into excel file
df.to_excel('/content/drive/MyDrive/Python_programming_project/python_dataset/employee.xlsx', index=False)

#close the connetion of sqlite3
conn.close()
```
